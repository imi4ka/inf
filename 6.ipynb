```python
from turtle import * # Импортируем черепаху
size=30 # Задаем масштаб окна (30 оптимальный но это интерактивное значение)
left(90) # Разварачиваем черепаху на 90°
screensize(2000,2000) # Задаем размер окна
tracer(0) # Отключаем анимацию что бы выполнить код "почти" моментально
down() # Опускает "хвост" чтобы начать рисовать (обязательно)

#
#
# Тут будет цикл условия
#
#

up() # Поднимает "хвост" что бы перестать рисовать (обязательно)

for x in range(-20,20):
    for y in range(-20,20):

        # Эти два вложенных цикла перебирают все возможные комбинации координат (x, y) в диапазоне от -20 до 19 
        # (поскольку range не включает конечное значение).
        # Таким образом, точки будут расположены в квадратной сетке с центром в (0, 0).

        setpos(x*size,y*size) # Перемещение в точку (x*size, y*size)
        dot(4, 'red')  # Рисование красной точки размером 4 пикселя
done() # Завершение работы с turtle
```

```python
from turtle import *
size=30
left(90)
screensize(2000,2000)
tracer(0)
down()

# Повтори 7 [Вперёд 10 Направо 120]
# Задаем стандартный цикл, по условию (радиус цикла - кол-во повторений)
# forward - вперёд (При использовании умножаем на size)
# left - влево
# right - вправо
# back - назад (При использовании умножаем на size)

for i in range(7): 
    forward(10*size)
    right(120)
    
up()

for x in range(-20,20):
    for y in range(-20,20):
        setpos(x*size,y*size)
        dot(4, 'red') 
done() 
```
