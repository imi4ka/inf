**Пример:** 

Для начала определеяем переменную s (строку):
- состоящей из 68 идущих подряд цифр 8

Получается s = "8" * 68

```python
s = "8" * 68
while "222" in s or "888" in s: # Запись цикла (Пока нашлось (222) или нашлось (888))
    if "222" in s:
        s = s.replace("222", "8", 1)
    else:
        s = s.replace("888", "2", 1)

print(s)
```

Пример друго трактовки строк: 

Другая трактовка: строка, состоящая из 55 цифр; последняя цифра в строке  — цифра 7, а остальные цифры  — пятёрки

Т.е строка s = "5" * 54 + "7"

```python
s = "5" * 54 + "7"
while "722" in s or "557" in s:
    if "722" in s:
        s = s.replace("722", "57", 1)
    else:
        s = s.replace("557", "72", 1)

print(s)
```

**Другой тип задания**
- 

Здесь будем к циклу While, добавлять цикл for для нахождения n

```python
for n in range(4, 10000): # Поле перебора 3 < n < 10000. Помним что счет начинается с 0
    s = "1" + "2" * n
    while "12" in s or "3322" in s or "2222" in s:
        if "12" in s:
            s = s.replace("12", "33", 1)
        if "2222" in s:
            s = s.replace("2222", "1", 1)
        if "3322" in s:
            s = s.replace("3322", "21", 1)
    
    d = sum(map(int, s))

    # map придает каждому числу интовое состояние
    # sum суммирует число

    if d == 218:
        print(n)
        break
```
