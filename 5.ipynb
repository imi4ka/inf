````python
for n in range(1, 1000):
    s = str(bin(n)[2:]) #       N
                        # [2:]     ('0b'),    

    if s.count('1') % 2 == 0: #   (     2  0, ,   ).
        s += str(s.count('1') % 2) #    s 0,    ,  1,  .
    r = int(s, 2) #     s     r.

    if r > 43:
        print(r, "|", s)
        break


    #     R > 43,       .
    #      ,       :
    # 44 = 101100
    # -  4 : 1011 (: 1+0+1+1 = 3  ).
    # -   2 : 10 ( ).
    # -   2 : 00   .

    # 45 = 101101
    # -  4 : 1011 (: 3  ).
    # -   2 : 10.
    # -   2 : 01   .

    # 46 = 101110
    # -  4 : 1011 (: 3  ).
    # -   2 : 10.
    # -   2 : 10  !
````

````python
for n in range(1, 1000):
    s = str(bin(n)[2:])

    if s.count('1') % 2 == 0:
        s += '00'
    else:
        s += '11'
    
    r = int(s, 2)

    if r > 114:
        print(r, '|', s)
        break
````

````python
#     x      
def s3(x):
    s = ''
    while x > 0:
        s = str(x % 3) + s
        x = x // 3
    return s

a = [] #   

for n in range(1, 1000):
    s = s3(n)
    
    if n % 3 == 0:
        s += s[-2:]
    else:
        s += s3(n % 3 * 3)

    r = int(s, 3)

    if r <= 150:
        a.append(n) #     N  

print(max(a)) #   N  
````
