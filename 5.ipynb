# Посимвольное двоичное преобразование



````python
for n in range(1, 1000):
    s = str(bin(n)[2:]) # Строим строку с двочиным представлением числа N
                        # [2:] обрезает первые два символа ('0b'), оставляя только двоичное представление

    if s.count('1') % 2 == 0: # проверяет чётность (если остаток от деления на 2 равен 0, значит, количество единиц чётное).
        s += str(s.count('1') % 2) # Добавляет к строке s 0, если количество единиц чётное, или 1, если нечётное.
    r = int(s, 2) # Преобразует полученную двоичную строку s обратно в десятичное число r.

    if r > 43:
        print(r, "|", s)
        break


    # Нам нужно найти минимальное R > 43, которое может получиться в результате работы алгоритма.
    # Рассмотрим двоичные представления чисел и проверим, соответствуют ли они правилу добавления двух битов:
    # 44 = 101100₂
    # - Первые 4 бита: 1011 (сумма: 1+0+1+1 = 3 → нечётная).
    # - Ожидаемые последние 2 бита: 10 (по алгоритму).
    # - Фактические последние 2 бита: 00 → не подходит.

    # 45 = 101101₂
    # - Первые 4 бита: 1011 (сумма: 3 → нечётная).
    # - Ожидаемые последние 2 бита: 10.
    # - Фактические последние 2 бита: 01 → не подходит.

    # 46 = 101110₂
    # - Первые 4 бита: 1011 (сумма: 3 → нечётная).
    # - Ожидаемые последние 2 бита: 10.
    # - Фактические последние 2 бита: 10 → подходит!
````

````python
for n in range(1, 1000):
    s = str(bin(n)[2:])

    if s.count('1') % 2 == 0:
        s += '00'
    else:
        s += '11'
    
    r = int(s, 2)

    if r > 114:
        print(r, '|', s)
        break
````

````python
# Эта функция переводит число x из десятичной системы счисления в троичную
def s3(x):
    s = ''
    while x > 0:
        s = str(x % 3) + s
        x = x // 3
    return s

a = [] # Задаем пустой список

for n in range(1, 1000):
    s = s3(n)
    
    if n % 3 == 0:
        s += s[-2:]
    else:
        s += s3(n % 3 * 3)

    r = int(s, 3)

    if r <= 150:
        a.append(n) # Заносим в список все N удовлетворяющие условие

print(max(a)) # Выводим максимальное N из списка
````
